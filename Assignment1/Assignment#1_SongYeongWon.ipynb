```python
# PassengerID, Name, Ticket 을 제거한 data 를 df라는 데이터 프레임으로 저장하였다. 
df = data.drop(['PassengerId', 'Name', 'Ticket'],axis='columns')
df
```

# 3. Data Analysis

## (c)

```python
# groupby(특정 컬럼)을 통해서 특정 컬럼을 그룹화 하여 묶을 수 있고 mean을 사용하여 그룹별로 묶었을 때 컬럼별 평균값을 확인 할 수 있다.
# 그 중에 성별로 생존률 평균을 확인해보면 아래와 같다.
df.groupby(['Sex']).mean().loc[:,['Survived']].reset_index()
```

## (d)

```python
# Age 열의 모든 값을 10으로 나누었을 때 정수부분만 구하여 나온 값을 새로운 age_group 이라는 연령대 열을 데이터 프레임에 추가하였다.
def func_age():
    df['Age']= (df['Age']//10*10).astype(int)
    df_age = df.groupby(['Age']).mean().loc[:,['Survived']].sort_values('Survived').reset_index()
    display(df_age)
```

```python
func_age()
```

## (e)

```python
age_sur=df[['Survived','Age','Sex']]
#df_survived 은 age_sur에서 비생존자일 경우의 추출한 데이터 프레임을 나타낸다.
df_nonsurvived=age_sur[age_sur['Survived'] == 0].reset_index(drop=True)
#df_nonsurvived 은 age_sur에서 생존자일 경우의 추출한 데이터 프레임을 나타낸다.
df_survived=age_sur[age_sur['Survived'] == 1].reset_index(drop=True)
#df1 은 df_survived 연령대별 비생존자의 수를 나타내는 데이터 프레임이다.
df1=df_nonsurvived.groupby(['Age']).count().reset_index()
#df1 은 df_survived 연령대별 생존자의 수를 나타내는 데이터 프레임이다.
df2=df_survived.groupby(['Age']).count().reset_index()
```

```python
plt.bar(df1['Age']+3, df1['Survived'],width=4)
plt.bar(df2['Age']+7, df2['Survived'],width=4)
plt.title("Age - Survival count",fontsize=12)
plt.xlabel("age")
plt.ylabel("count")
plt.legend(labels = ['Survived X','Survived O'],loc = 'upper right')
plt.xlim(0,90)
plt.grid(True)
```

## (f)

```python
#df_survived_male 은 df_survived 에서 성별이 male 인 경우만 추출한 데이터 프레임을 나타낸다.
df_survived_male=df_survived[df_survived['Sex'] == 'male'].reset_index(drop=True)

#df_nonsurvived_male 은 df_nonsurvived 에서 성별이 male 인 경우만 추출한 데이터 프레임을 나타낸다.
df_nonsurvived_male=df_nonsurvived[df_nonsurvived['Sex'] == 'male'].reset_index(drop=True) 

#df3 은 df_nonsurvived_male 에서 연령대별 비생존자의 수를 나타내는 데이터 프레임이다.
df3=df_nonsurvived_male.groupby(['Age']).count().reset_index()

#df4 은 df_survived_male 에서 연령대별 생존자의 수를 나타내는 데이터 프레임이다.
df4=df_survived_male.groupby(['Age']).count().reset_index()

#df_survived_female 은 df_survived 에서 성별이 female 인 경우만 추출한 데이터 프레임을 나타낸다.
df_survived_female=df_survived[df_survived['Sex'] == 'female'].reset_index(drop=True)

#df_nonsurvived_female 은 df_nonsurvived 에서 성별이 female 인 경우만 추출한 데이터 프레임을 나타낸다.
df_nonsurvived_female=df_nonsurvived[df_nonsurvived['Sex'] == 'female'].reset_index(drop=True)

#df5 은 df_nonsurvived_female 에서 연령대별 비생존자의 수를 나타내는 데이터 프레임이다.
df5=df_nonsurvived_female.groupby(['Age']).count().reset_index()

#df6 은 df_survived_male 에서 연령대별 생존자의 수를 나타내는 데이터 프레임이다.
df6=df_survived_female.groupby(['Age']).count().reset_index()
```

```python
f, (ax1, ax2) = plt.subplots(1, 2,figsize=(15,5))

#각각 df_male1,df_male2 의 Age 열을 히스토그램으로 나타내면 아래와 같이 male 일 경우 나이별 사망자와 생존자 분포를 확인 할 수 있다.
ax1.bar(df3['Age']+3, df3['Survived'],width=4)
ax1.bar(df4['Age']+7, df4['Survived'],width=4)
ax1.set_title("Male",fontsize=12)
ax1.set_xlabel("age")
ax1.set_ylabel("count")
ax1.legend(labels = ['Survived X','Survived O'],loc = 'upper right')
ax1.set_xlim(0,90)
ax1.set_ylim(0,130)
ax1.grid(True)
#각각 df_female1,df_female2 의 Age 열을 히스토그램으로 나타내면 아래와 같이 female 일 경우 나이별 사망자와 생존자 분포를 확인 할 수 있다.
ax2.bar(df5['Age']+3, df5['Survived'],width=4)
ax2.bar(df6['Age']+7, df6['Survived'],width=4)
ax2.set_title("Female",fontsize=12)
ax2.set_xlabel("age")
ax2.set_ylabel("count")
ax2.legend(labels = ['Survived X','Survived O'],loc = 'upper right')
ax2.set_xlim(0,90)
ax2.set_ylim(0,130)
ax2.grid(True)
```

# Extra analysis

case 1) 동승자 수에 따른 생존률 분석

```python
#추가적인 데이터 분석에서는 생존에 영향을 미칠 수 있는 변수로 Sibsp 와 Parch 를 정하여서 분석하였다.
#Sibsp 와 Parch을 각각 동승자 수에 따른 생존과 비생존의 수를 확인 하고자 한다.
```

```python
df_SibSp=df[['Survived','SibSp']]
df_Parch=df[['Survived','Parch']]
df_Parch
```

```python
df_SibSp.groupby(by = ['Survived','SibSp'])['Survived'].count()
df_Parch.groupby(by = ['Survived','Parch'])['Survived'].count()
```

```python
#Sibsp 와 Parch을 각각 동승자 수에 따른 생존과 비생존자의 수를 countplot 을 통해 그래프로 확인하였다.
#Sibsp 와 Parch 모두 동승자가 없이 단독인 경우 가장 많은 사람이 생존하지 못하였음을 쉽게 확인 할 수 있다.
f, (ax3, ax4) = plt.subplots(1, 2,figsize=(10,5))
sns.countplot(data = df_SibSp, x='SibSp',hue='Survived',ax=ax3)
ax3.legend(labels = ['Survived : X','Survived : O'],loc = 'upper right')
sns.countplot(data = df_Parch, x='Parch',hue='Survived',ax=ax4)
ax4.legend(labels = ['Survived : X','Survived : O'],loc = 'upper right')
```

```python
#Sibsp 와 Parch 의 모두 동승자 수에 따른 평균생존률을 구하였다.
SibSp_sur=df.groupby(['SibSp']).mean().loc[:,['Survived']].reset_index()
Parch_sur=df.groupby(['Parch']).mean().loc[:,['Survived']].reset_index()
Parch_sur
```

```python
#Sibsp 와 Parch 의 동승자별 생존자 평균을 확인하면 아래와 같다.
#동승자가 없는 경우 가장 많은 사망자가 발생하였지만, 동승자가 없는 경우 생존자 또한 가장 많기 때문에 전체 생존자 중에 동승자가 없는 경우
#의 생존률은 동승자가 있는 경우에 비해 크게 차이 나지 않는다.

SibSp_series = pd.Series(SibSp_sur['Survived'], index = SibSp_sur['SibSp'])
Parch_series = pd.Series(Parch_sur['Survived'], index = Parch_sur['Parch'])
explode_SibSp = pd.Series([0.05, 0.05,0.05, 0.05, 0.05,0.05], index = SibSp_series.index)
explode_Parch = pd.Series([0.05, 0.05,0.05, 0.05, 0.05,0.05,0.05], index = Parch_series.index)
f, (ax4, ax5) = plt.subplots(1, 2,figsize=(10,5))
ax4.pie(SibSp_sur['Survived'],labels = SibSp_sur['SibSp'], autopct = '%0.1f%%',explode = explode_SibSp)
ax4.set_title("Survived-SibSp")
ax5.pie(Parch_sur['Survived'],labels = Parch_sur['Parch'], autopct = '%0.1f%%',explode = explode_Parch)
ax5.set_title("Survived-Parch")
```

case 2) 동승자 유무에 따른 생존률 분석

```python
# Sibsp_our 이라는 열을 새로 추가하여 Sibsp 의 동승자가 있을 경우 Not_alone 을 없을 경우 alone 으로 표기하였다.
df['Sibsp_our']= ["Not_alone" if i >0 else "alone" for i in df['SibSp']] 
# Parch_our 이라는 열을 새로 추가하여 Parch 의 동승자가 있을 경우 Not_alone 을 없을 경우 alone 으로 표기하였다.
df['Parch_our']= ["Not_alone" if i >0 else "alone" for i in df['Parch']] 
df
```

```python
df_Sibsp_our=df[['Survived','Sibsp_our']]
df_Parch_our=df[['Survived','Parch_our']]
df_Parch_our
```

```python
#Sibsp 와 Parch을 각각 동승자 유무에 따른 생존과 비생존자의 수를 countplot 을 통해 그래프로 확인하였다.
#아래의 그래프를 통해서 Sibsp 와 Parch 모두 동승자가 있는 경우는 생존자와 사망자의 수 차이가 작다.
#하지만 동승자가 없는 경우는 생존자의 비해 사망자가 훨씬 많음을 확인 할 수 있다.
f, (ax3, ax4) = plt.subplots(1, 2,figsize=(10,5))
sns.countplot(data = df_Sibsp_our, x='Sibsp_our',hue='Survived',ax=ax3)
ax3.legend(labels = ['Survived : X','Survived : O'],loc = 'upper right')
sns.countplot(data = df_Parch_our, x='Parch_our',hue='Survived',ax=ax4)
ax4.legend(labels = ['Survived : X','Survived : O'],loc = 'upper right')
```

```python
#Sibsp 와 Parch 의 모두 동승자 수에 따른 평균생존률을 구하였다.
#아래 수치를 통해 확인해보면, 동승자가 있는 경우 보다 동승자가 없는 경우 사망률이 더 높게 난다.
#따라서 동승자의 수 보다 동승자의 유무에 따라 생존에 대해 분석하는 것이 더 적절하다고 판단된다.
Sibsp_our=df.groupby(['Sibsp_our']).mean().loc[:,['Survived']].reset_index()
Parch_our=df.groupby(['Parch_our']).mean().loc[:,['Survived']].reset_index()
```
